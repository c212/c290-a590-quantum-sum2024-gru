```python
q = QuantumCircuit(3)
q.h(0)
q.cx(0, 1)
q.cx(0, 2)
q.barrier()
q.sdg(1)
q.sdg(2)
q.barrier()
q.h(0)
q.h(1)
q.h(2)
# q.measure_all()
q.draw(output='mpl')
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_instruction(q)
sv.draw('latex')
```
